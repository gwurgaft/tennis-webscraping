```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
urls = {"Ben Shelton": "https://www.tennisabstract.com/cgi-bin/player.cgi?p=BenShelton",
       "Carlos Alcaraz": "https://www.tennisabstract.com/cgi-bin/player.cgi?p=CarlosAlcaraz",
       "Jannik Sinner": "https://www.tennisabstract.com/cgi-bin/player.cgi?p=JannikSinner"}
driver = webdriver.Chrome()
total_tables = {}
for i in urls:
    driver.get(urls[i])
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    tables = soup.find_all("table")
    total_tables[i] = tables
driver.quit()
print(tables)
# driver.get("https://www.tennisabstract.com/cgi-bin/player.cgi?p=BenShelton")
# html_shelton = driver.page_source
# soup_shelton = BeautifulSoup(html_shelton, "html.parser")
# tables = soup_shelton.find_all("table")

# driver.get("https://www.tennisabstract.com/cgi-bin/player.cgi?p=CarlosAlcaraz")
# html_alcaraz = driver.page_source
# soup = BeautifulSoup(html_alcaraz, "html.parser")

# driver.quit()
```

```python
for j in total_tables:
    dfs = []
    for idx, table in enumerate(tables, start=1):
        headers = [th.get_text(strip=True) for th in table.find_all("th")]
        rows = []
        for tr in table.find_all("tr")[1:]:
            cells = [td.get_text(strip=True) for td in tr.find_all("td")]
            if cells:
                rows.append(dict(zip(headers, cells)))
        df = pd.DataFrame(rows)
        dfs.append((f"table_{idx}", df))
    complete_dfs = []
    for df in dfs:
        if df[1].empty != True:
            complete_dfs.append(df)
    total_tables[j] = complete_dfs
    folder_name = j
    os.mkdir(folder_name)
    print(f"Folder '{folder_name}' created.")
```
